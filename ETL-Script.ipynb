```python
#-----------Establishing Connection------------
cnxn_str = (
    'DRIVER={ODBC Driver 18 for SQL Server};'
    'SERVER=tcp:dfv1.database.windows.net,1433;'
    'DATABASE=DFT1;'
    'UID=digifloat_training_db;'
    'PWD=Hitter@777'
)

conn = pyodbc.connect(cnxn_str)
print("Connection succesful")

#-----------Creating a Table------------
conn.execute("""
CREATE TABLE Aftab (
    id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    email VARCHAR(100),
    age INT,
    city VARCHAR(50)
);
""")
conn.commit()
print("its done")
conn.close()
print("connetion closed")
```

```python
#--------Fecthing Records from CSV File--------------
df3 = pd.read_csv('user.csv')
df3.head(10)
```

```python
#-----------Inserting Records into Table-------------
conn = pyodbc.connect(cnxn_str)
print("Connection succesful")
id = 0
for idx,row in df3.iterrows():
    conn.execute(f"""
        INSERT INTO Aftab (id, first_name, last_name, email, age, city) VALUES
        ({id},'{row.first_name}','{row.last_name}','{row.email}',{int(row.age)},'{row.city}')
        """)
    id += 1

print("Successfully Inserted to SQL Table")
conn.commit()
print("its done")
conn.close()
print("connetion closed")

```

```python
#---------Fetching from Cloud SQL server-----------
conn = pyodbc.connect(cnxn_str)
print("Connection succesful")
df = pd.read_sql_query("""
select * from Aftab
""",conn)
conn.commit()
print("its done")
conn.close()
print("connetion closed")

df.head(10)
```

```python
#---------Fetching from JSON File-----------
df2 = pd.read_json('user.json')
df2.head(10)
```

```python
#---------Merging Data of All files-----------
df0 = pd.concat([df, df2, df3])
df0.head(50)
```

```python
#----------Storing Resultant Schema With Data-----------
df0.to_json('myData.json', orient='table', indent=4, index=False)
print("Successful Json File Created")
```

```python
#----------Storing Resultant Schema Only-----------
import json
# Extract schema information
schema = {
    'columns': list(df0.columns),
    'dtypes': df0.dtypes.astype(str).to_dict()
}

# Save schema to JSON file
with open('schema.json', 'w') as f:
    json.dump(schema, f, indent=4)

print("Successful Json File Created")
```
